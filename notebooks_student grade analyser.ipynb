```python
# Task 1
def letter_grade(score):
    if score < 0 and score > 100:
        raise ValueError(f"Invalid score: {score} must be between 0 and 100")

    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

    
```

```python
#Task 2 
def analyse(students):
    analysis_results = {}                                                          #dictionary
    for name, score in students.items():
       grade = letter_grade(score)                                                     #calling task 1 function       
       analysis_results[name] = {"score": score, "grade": grade}                        # storing the nested dict
    return analysis_results                                                            
```

```python
#Task 3
def summary(results):
    for name,data in results.items():
        print(f"{name:<7}: {data['score']} → {data['grade']} ")
    print()                                                                     #a blank line

    #class ko average
    total_score = sum(data['score'] for data in results.values())
    average = total_score/len(results)

    #highest and lowest
    highest_student = max(results, key = lambda k:results[k] ["score"])
    lowest_student = min(results, key = lambda k:results[k] ["score"])

    highest_score = results[highest_student]["score"]
    lowest_score = results[lowest_student]["score"]

    #grade kati choti occur bhayo
    grade_counts = {}
    for data in results.values():
        g = data["grade"]
        grade_counts[g] = grade_counts.get(g, 0) + 1

   
    all_possible_grades = ["A", "B", "C", "D", "F"]
    counts_str = "  ".join([f"{g}={grade_counts.get(g, 0)}" for g in all_possible_grades])

    # Summary Statistics
    print(f"Class Average : {average:.2f}")
    print(f"Highest Score : {highest_student} ({highest_score})")
    print(f"Lowest Score  : {lowest_student} ({lowest_score})")
    print(f"Grade Counts  : {counts_str}")
```

```python
#Task 4
valid_students = {
    "Alice": 92, "Bob": 78, "Carol": 85,
    "Dave": 61, "Eve": 55, "Frank": 99
}

invalid_students = {
    "Alice": 92, "Ghost_Student": 110, "Bob": 78
}

print("-----Running Valid Data-----")
try:
    processed_results = analyse(valid_students)
    summary(processed_results)
except ValueError as e:
    print(f"An error occurred during processing: {e}")

print("\n" + "="*30 + "\n")

print("------Running Invalid Data------")
try:
    processed_results = analyse(invalid_students)
    summary(processed_results)
except ValueError as e:
    print(f"Caught expected error gracefully: {e}")
```
