```python
# Task 3:
#Borrow a book
def borrow(self):
    if not self.available:
        raise ValueError(f"'{self.title}' is already borrowed.")
    self.available = False
        
# Return a book
def return_book(self):
    if self.available:
        raise ValueError(f"'{self.title}' is not currently borrowed.")
    self.available = True
    
```

```python
# Task 4:
#string representation
def __str__(self):
    status = "✓" if self.available else "✗"
    return f"[{status}] {self.title} | {self.author} | {self.genre}"
```

```python
# submisssion file for this 
class Book:
    total_books = 0

    def __init__(self, title, author, genre):
        self.title = title
        self.author = author
        self.genre = genre
        self.available = True  
        Book.total_books += 1

    @classmethod
    def from_dict(cls, data):
        return cls(data["title"], data["author"], data["genre"])

    @classmethod
    def get_total(cls):
        return f"Total books registered: {cls.total_books}"

    def borrow(self):
        if not self.available:
            raise ValueError(f"'{self.title}' is already borrowed.")
        self.available = False

    def return_book(self):
        if self.available:
            raise ValueError(f"'{self.title}' is not currently borrowed.")
        self.available = True

    def __str__(self):
        status = "✓" if self.available else "✗"
        return f"[{status}] {self.title} | {self.author} | {self.genre}"


b1 = Book("Python Crash Course", "Eric Matthes", "Programming")
b2 = Book("Sapiens", "Yuval Noah Harari", "History")
b3 = Book.from_dict({
    "title": "Deep Work", "author": "Cal Newport", "genre": "Productivity"
})

b1.borrow()
b2.borrow()

print(b1)  
print(b2)  
print(b3)  

print()

b1.return_book()
print(b1)  

print()

try:
    b2.borrow()  
except ValueError as e:
    print(f"❌ ValueError: {e}")

print()
print(Book.get_total())
```
