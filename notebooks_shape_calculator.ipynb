```python
#task 1:
import math

class Circle:
    shapes_created = 0

    def __init__(self,radius):
        if not self.is_valid_dimension(radius):                      #number must be positive.validation
            raise ValueError("radius must be positive.")

        self.radius = radius
        Circle.shapes_created += 1

    def area(self):
        return round(math.pi * (self.radius ** 2), 2)

    def perimeter(self):
        return round(math.pi * (self.radius ** 2), 2)

    def __str__(self):
        return f"Circle(r={self.radius})"
```

```python
#task 2:
class Rectangle:
    shapes_created = 0  

    def __init__(self, width, height):
        # Both numbers must be positive
        if not self.is_valid_dimension(width) or not self.is_valid_dimension(height):
            raise ValueError("dimensions must be positive")
            
        self.width = width
        self.height = height
        Rectangle.shapes_created += 1  
        
    def area(self):
        # area of rect =  width × height
        return round(self.width * self.height, 2)

    def perimeter(self):
        # perimeter of rect =  2 × (width + height)
        return round(2 * (self.width + self.height), 2)

    def is_square(self):
        # check if it a square 
        return self.width == self.height

    def __str__(self):
        return f"Rectangle({self.width} x {self.height})"
```

```python
#task 3:
@staticmethod
def cm_to_inch(cm):
    return round(cm * 0.3937, 3)                           # Formula: cm × 0.3937
    
@staticmethod
def is_valid_dimension(val):
    return val > 0
    
```
