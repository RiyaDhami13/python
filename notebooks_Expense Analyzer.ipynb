```python
# task 1
def load_expenses(filename):
    expenses = []
    try:
        with open(filename, "r") as file:
            for line in file:
                cleaned_line = line.strip()

                if not cleaned_line:  # skipping the empty line
                    continue

                if ',' not in cleaned_line:
                    continue

                category, amount_str = cleaned_line.split(',')

                expenses.append((category, float(amount_str)))  # amount to float and storing as tuple            

    except FileNotFoundError:  # kehi na bhaye...empty return garni 
        return []

    return expenses
                            
```

```python
#task 2
def add_expense(filename, category, amount):
    if amount <= 0:
        raise ValueError(f"Invalid amount: amount must be greater than 0")

    # Appending to the file:
    with open(filename,"a") as file:
        file.write(f"{category},{amount}\n")
```

```python
#task 3
def category_totals(expenses):
    totals = {}
    grand_total = 0.0
    
    # Group and sum spending per category
    for category, amount in expenses:
        totals[category] = totals.get(category, 0.0) + amount
        grand_total += amount                                     
        

    print(f"{'Category':<16} Total")
    print("─" * 23)
    
    
    for category, total in totals.items():
        print(f"{category:<15} : ${total:.2f}")
        
    # 4. Print the grand total
    print("─" * 23)
    print(f"{'Grand Total':<15} : ${grand_total:.2f}")
```

```python
# task 4
def above_threshold(expenses, limit):
    #list comprehension to filter tuples 
    return [(category, amount) for category, amount in expenses if amount > limit]
```
